```python
# Step 1: Create one long list where each row contains team and match_id
team_matches = pd.concat([
    df[['Match_ID', 'team_1']].rename(columns={'team_1': 'Team_demo'}),
    df[['Match_ID', 'team_2']].rename(columns={'team_2': 'Team_demo'})
])

# Step 2: Drop duplicates to avoid counting the same match multiple times for the same team
team_matches = team_matches.drop_duplicates()

# Step 3: Group and count
matches_per_team = team_matches.groupby('Team_demo')['Match_ID'].nunique().sort_values(ascending=False)

print(matches_per_team)
```
